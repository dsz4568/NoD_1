```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generowanie przykładowych danych
np.random.seed(42)  # Ustawienie ziarna dla powtarzalności wyników

# Generowanie danych dla sklepu A
sprzedaz_A = np.random.normal(100, 20, 30)  # Średnia sprzedaż 100, odchylenie standardowe 20
cena_A = np.random.normal(50, 10, 30)  # Średnia cena 50, odchylenie standardowe 10

# Generowanie danych dla sklepu B
sprzedaz_B = np.random.normal(120, 25, 30)  # Średnia sprzedaż 120, odchylenie standardowe 25
cena_B = np.random.normal(55, 12, 30)  # Średnia cena 55, odchylenie standardowe 12

# Tworzenie DataFrame
dane = pd.DataFrame({
    'Sprzedaz_A': sprzedaz_A,
    'Cena_A': cena_A,
    'Sprzedaz_B': sprzedaz_B,
    'Cena_B': cena_B
})
```

```python
# Obliczanie średniej sprzedaży w obu sklepach
srednia_sprzedaz_A = dane['Sprzedaz_A'].mean()
srednia_sprzedaz_B = dane['Sprzedaz_B'].mean()

print(f"Średnia sprzedaż w sklepie A: {srednia_sprzedaz_A:.2f}")
print(f"Średnia sprzedaż w sklepie B: {srednia_sprzedaz_B:.2f}")
```

```python
# Obliczanie mediany sprzedaży w obu sklepach
mediana_sprzedaz_A = dane['Sprzedaz_A'].median()
mediana_sprzedaz_B = dane['Sprzedaz_B'].median()

print(f"Mediana sprzedaży w sklepie A: {mediana_sprzedaz_A:.2f}")
print(f"Mediana sprzedaży w sklepie B: {mediana_sprzedaz_B:.2f}")
```

```python
# Obliczanie odchylenia standardowego sprzedaży w obu sklepach
odchylenie_sprzedaz_A = dane['Sprzedaz_A'].std()
odchylenie_sprzedaz_B = dane['Sprzedaz_B'].std()

print(f"Odchylenie standardowe sprzedaży w sklepie A: {odchylenie_sprzedaz_A:.2f}")
print(f"Odchylenie standardowe sprzedaży w sklepie B: {odchylenie_sprzedaz_B:.2f}")
```

```python
# Obliczanie korelacji między ceną a sprzedażą w obu sklepach
korelacja_A = dane['Cena_A'].corr(dane['Sprzedaz_A'])
korelacja_B = dane['Cena_B'].corr(dane['Sprzedaz_B'])

print(f"Korelacja między ceną a sprzedażą w sklepie A: {korelacja_A:.2f}")
print(f"Korelacja między ceną a sprzedażą w sklepie B: {korelacja_B:.2f}")
```

```python
# Wizualizacja danych
plt.figure(figsize=(12, 6))

# Wykres sprzedaży w sklepie A
plt.subplot(1, 2, 1)
sns.scatterplot(x='Cena_A', y='Sprzedaz_A', data=dane)
plt.title('Sklep A: Cena vs Sprzedaż')
plt.xlabel('Cena')
plt.ylabel('Sprzedaż')

# Wykres sprzedaży w sklepie B
plt.subplot(1, 2, 2)
sns.scatterplot(x='Cena_B', y='Sprzedaz_B', data=dane)
plt.title('Sklep B: Cena vs Sprzedaż')
plt.xlabel('Cena')
plt.ylabel('Sprzedaż')

plt.tight_layout()
plt.show()
```
