```python
# 1. Ładowanie biblioteki Pandas
import pandas as pd
```

```python
# 2. Tworzenie ramki danych ze słownika
data = {
    'Imię': ['Anna', 'Jan', 'Maria', 'Piotr'],
    'Wiek': [25, 30, 22, 35],
    'Miasto': ['Warszawa', 'Kraków', 'Gdańsk', 'Wrocław']
}
df_dict = pd.DataFrame(data)
print("Ramka danych ze słownika:")
print(df_dict)
```

```python
# 3. Zachowanie ramki danych pobranych z pliku w formacie CSV (lub XLSX)
# Zakładamy, że plik CSV nazywa się 'dane.csv'
df = pd.read_csv('IHME_HEALTH_SPENDING_1995_2018_Y2021M09D22.csv')
df.to_excel('dane.xlsx', index=False)  # Zapis do pliku Excel
print("\nRamka danych zapisana do pliku Excel.")
```

```python
# 4. Tworzenie ramki danych z listy list
data_list = [['Anna', 25, 'Warszawa'], ['Jan', 30, 'Kraków'], ['Maria', 22, 'Gdańsk']]
df_list = pd.DataFrame(data_list, columns=['Imię', 'Wiek', 'Miasto'])
print("\nRamka danych z listy list:")
print(df_list)
```

```python
# 5. Transponowanie (wymiana kolumn i wierszy)
df_transposed = df.transpose()
print("\nTransponowana ramka danych:")
print(df_transposed)
```

```python
# 6. Wyświetlenie pierwszych 10 wierszy ramki danych
print("\nPierwsze 10 wierszy ramki danych:")
print(df.head(10))
```

```python
# 7. Wyświetlenie ostatnich 10 wierszy ramki danych
print("\nOstatnie 10 wierszy ramki danych:")
print(df.tail(10))
```

```python
# 8. Wyświetlenie informacji o ramce danych
print("\nInformacje o ramce danych:")
print(df.info())
```

```python
# 9. Wyświetlenie liczby wierszy i kolumn w ramce danych
print("\nLiczba wierszy i kolumn:")
print(df.shape)
```

```python
# 10. Wyświetlenie informacji statystycznych o kolumnach liczbowych
print("\nStatystyki kolumn liczbowych:")
print(df.describe())
```

```python
# 11. Wyświetlenie informacji statystycznych o kolumnach kategoryzowanych
print("\nStatystyki kolumn kategoryzowanych:")
print(df.describe(include=['object']))
```

```python
# 12. Usunięcie brakujących wartości w ramce danych
df_cleaned = df.dropna()
print("\nRamka danych po usunięciu brakujących wartości:")
print(df_cleaned)
```

```python
# 13. Wybór wierszy i kolumn za pomocą nazw oraz indeksów
print("\nWybór wierszy i kolumn:")
print(df.loc[0:2, ['location_name', 'iso3']])  # Użycie nazw kolumn
print(df.iloc[0:2, 0:2])  # Użycie indeksów
```

```python
# 14. Wybór wierszy pod warunkiem określonej wartości kolumny
print("\nWybór wierszy, gdzie rok > 1995:")
print(df[df['year'] > 25])
```

```python
# 15. Wybór wierszy pod warunkiem spełnienia kilku warunków jednocześnie
print("\nWybór wierszy, gdzie rok > 2000 i lokalizacja == 'Angola':")
print(df[(df['year'] > 25) & (df['location_name'] == 'Angola')])
```

```python
# 16. Wybór wierszy zawierających określone słowo w kolumnie kategoryzowanej
print("\nWybór wierszy, gdzie lokalizacja zawiera 'Anti':")
print(df[df['location_name'].str.contains('Anti')])
```

```python
# 17. Wybór wierszy, które nie zawierają określonego słowa w kolumnie kategoryzowanej
print("\nWybór wierszy, gdzie lokalizacja nie zawiera 'Angola':")
print(df[~df['location_name'].str.contains('Warszawa')])
```

```python
# 18. Utworzenie nowej kolumny na podstawie istniejącej
df['ppp_total_mean_plus_5'] = df['ppp_total_mean'] + 5
print("\nRamka danych z nową kolumną 'ppp_total_mean_5':")
print(df)
```

```python
# 19. Usunięcie kolumny
df = df.drop(columns=['ppp_total_mean_plus_5'])
print("\nRamka danych po usunięciu kolumny 'ppp_total_mean__plus_5':")
print(df)
```

```python
# 20. Zmiana nazwy kolumny
df = df.rename(columns={'year': 'rok'})
print("\nRamka danych po zmianie nazwy kolumny 'year' na 'rok':")
print(df)
```

```python
# 21. Zapisanie ramki danych jako plik CSV
df.to_csv('dane_modified.csv', index=False)
print("\nRamka danych zapisana jako plik CSV.")
```

```python
# 22. Wyświetlenie średniej, maksymalnej i minimalnej wartości z jednej kolumny
print("\nŚrednia wartość kolumny 'the_total_mean':", df['the_total_mean'].mean())
print("Maksymalna wartość kolumny 'the_total_mean':", df['the_total_mean'].max())
print("Minimalna wartość kolumny 'the_total_mean':", df['the_total_mean'].min())
```

```python
# 23. Wyświetlenie liczby wierszy
print("\nLiczba wierszy:", len(df))
```

```python
# 24. Wyświetlenie wartości unikatowych w kolumnie
print("\nUnikatowe wartości w kolumnie 'iso3':")
print(df['iso3'].unique())
```

```python
# 25. Wyświetlenie liczby rekordów odpowiadających wartościom
print("\nLiczba rekordów dla każdej wartości w kolumnie 'iso3':")
print(df['iso3'].value_counts())
```

```python
# 26. Sortowanie wierszy według wartości określonej kolumny (malejąco, rosnąco)
print("\nSortowanie wierszy według kolumny 'the_total_mean' (rosnąco):")
print(df.sort_values(by='the_total_mean', ascending=True))
print("\nSortowanie wierszy według kolumny 'the_total_mean' (malejąco):")
print(df.sort_values(by='the_total_mean', ascending=False))
```

```python
# 27. Wyświetlenie 10 największych (najmniejszych) wartości określonej kolumny
print("\n10 największych wartości kolumny 'the_total_mean':")
print(df.nlargest(10, 'the_total_mean'))
print("\n10 najmniejszych wartości kolumny 'the_total_mean':")
print(df.nsmallest(10, 'the_total_mean'))
```

```python
# 28. Wyświetlenie 10 największych wartości kolumny pod warunkiem określonych wartości innej kolumny
print("\n10 największych wartości kolumny 'the_total_mean' dla lokalizacji 'Afghanistan':")
print(df[df['location_name'] == 'Afghanistan'].nlargest(10, 'the_total_mean'))
```

```python
# 29. Grupowanie wierszy według wartości kolumny kategoryzowanej i uśrednienie wartości
print("\nGrupowanie według kolumny 'location_name' i uśrednienie wartości:")
print(df.groupby('location_name').mean(numeric_only=True))
```

```python
# 30. Grupowanie wierszy i obliczenie średniej, liczby wartości i mediany
print("\nGrupowanie i obliczenia statystyczne:")
print(df.groupby('the_total_ppp_mean').agg({'the_total_mean': ['mean', 'count', 'median']}))
```

```python
# 31. Wyświetlenie nazw kolumn indeksu złożonego
df_multiindex = df.set_index(['location_name', 'the_total_mean'])
print("\nNazwy kolumn indeksu złożonego:")
print(df_multiindex.index.names)
```

```python
# 32. Sortowanie kolumny indeksu złożonego
print("\nSortowanie indeksu złożonego:")
print(df_multiindex.sort_index(level='location_name'))
```

```python
# 33. Tworzenie tabeli przestawnej (pivot table)
print("\nTabela przestawna:")
pivot_table = df.pivot_table(values='the_total_mean', index='location_id', columns='iso3', aggfunc='mean')
print(pivot_table)
```

```python
# 34. Wyświetlenie indeksów i kolumn tabeli przestawnej
print("\nIndeksy tabeli przestawnej:", pivot_table.index)
print("Kolumny tabeli przestawnej:", pivot_table.columns)
```

```python
# 35. Utworzenie indeksu złożonego tabeli przestawnej i wyświetlenie go
pivot_table_multiindex = pivot_table.stack()
print("\nIndeks złożony tabeli przestawnej:")
print(pivot_table_multiindex)
```

```python
# 40. Łączenie ramek danych za pomocą metod merge i concat
df1 = pd.DataFrame({'A': ['A0', 'A1'], 'B': ['B0', 'B1']})
df2 = pd.DataFrame({'A': ['A2', 'A3'], 'B': ['B2', 'B3']})
print("\nŁączenie ramek danych metodą concat:")
print(pd.concat([df1, df2]))
print("\nŁączenie ramek danych metodą merge:")
print(pd.merge(df1, df2, on='A', how='outer'))
```

```python
# 41. Dodawanie nowych kolumn za pomocą operacji matematycznych
df['the_total_mean_squared'] = df['the_total_mean'] ** 2
print("\nRamka danych z nową kolumną 'the_total_mean_squared':")
print(df)
```

```python
# 42. Dodawanie nowych kolumn za pomocą funkcji lambda
df['the_total_mean_doubled'] = df['the_total_mean'].apply(lambda x: x * 2)
print("\nRamka danych z nową kolumną 'the_total_mean_doubled':")
print(df)
```
